```python
import pandas as pd
data = pd.read_csv('HW 5 data 13.4.csv', sep=',', na_values=".")
data
```

```python
#13.4  A process engineer is trying to improve the life of a cutting tool. He has run a 23 experiment using cutting speed (A), metal hardness (B), and cutting angle (C) as the factors.
```

```python
df = data
df.info()
```

```python
X = df[['a', 'b',  'c',]]
X
```

```python
xx=X.to_numpy()
poly= PolynomialFeatures(degree=3, interaction_only=True)
xxpd=pd.DataFrame(poly.fit_transform(xx), columns=['#1', 'a', 'b', 'c','ab', 'ac', 'bc', 'abc'])
xxpd
```

```python
Y=df[['Totals']]
regr=LinearRegression()
regr.fit(X,Y)
Yhat=regr.predict(X)
Yhat
```

```python
X = sm.add_constant(X)
X
```

```python
model = sm.OLS(Y, X)
res=model.fit()
print(res.summary())
```

```python
# a) Do any of the three factors affect tool life?

# Yes, cutting speed and cutting angle have a significant interaction which would affect the tool life.
```

```python
# b) What combination of factor levels produces the longest tool life?

# The longest tool life would be the factor levels bc. 
```

```python
# c) Is there a combination of cutting speed and cutting angle that always gives good results regardless of metal hardness?

# The best combination for good results regardless of metal hardness would be low cutting speed (a) and high cutting angle (c)
```
